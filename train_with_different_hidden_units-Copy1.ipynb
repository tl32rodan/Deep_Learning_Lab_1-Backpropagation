```python
a = np.array([0.8,0.2])
b = np.array([1,0])
-np.sum(b*np.log2(a))
```

```python
def forward( y, ground_truth):
        if y.ndim == 1:
            ground_truth = ground_truth.reshape(1, ground_truth.size)
            y = y.reshape(1, y.size)
            
        batch_size = y.shape[0]
        delta = 1e-7 # To avoid -inf
        return -np.sum(ground_truth * np.log2(y + delta)) / batch_size
```

```python
forward(a,b)
```
